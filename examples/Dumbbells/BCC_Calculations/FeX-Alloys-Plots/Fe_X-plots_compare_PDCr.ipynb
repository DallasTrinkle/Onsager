## Fe-X alloy plots
Here, we plot the partial diffusion coefficient ratios from our calculations and compare with the SCMF results from Messina et. al.

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File("../Fe-Cu_Ni_Si/Si_data.h5","r") as fl:
    Temp = np.array(fl['Temp'])
    diff_aa_Si = np.array(fl['diff_aa'])
    diff_ab_Si = np.array(fl['diff_ab'])
    diff_bb_Si = np.array(fl['diff_bb'])
    diff_bb_Si_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("../Fe-Cu_Ni_Si/Cu_data.h5","r") as fl:
    diff_aa_Cu = np.array(fl['diff_aa'])
    diff_ab_Cu = np.array(fl['diff_ab'])
    diff_bb_Cu = np.array(fl['diff_bb'])
    diff_bb_Cu_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("../Fe-Mn/Mn_data.h5","r") as fl:
    diff_aa_Mn = np.array(fl['diff_aa'])
    diff_ab_Mn = np.array(fl['diff_ab'])
    diff_bb_Mn = np.array(fl['diff_bb'])
    diff_bb_Mn_non_local = np.array(fl['diff_bb_nl'])
    
with h5py.File("../Fe-Cu_Ni_Si/Ni_data.h5","r") as fl:
    diff_aa_Ni = np.array(fl['diff_aa'])
    diff_ab_Ni = np.array(fl['diff_ab'])
    diff_bb_Ni = np.array(fl['diff_bb'])
    diff_bb_Ni_non_local = np.array(fl['diff_bb_nl'])
    
with h5py.File("../Fe-X-60-only/Cr_data.h5","r") as fl:
    diff_aa_Cr = np.array(fl['diff_aa'])
    diff_ab_Cr = np.array(fl['diff_ab'])
    diff_bb_Cr = np.array(fl['diff_bb'])
    diff_bb_Cr_non_local = np.array(fl['diff_bb_nl'])
```

```python
# For Cu, Ni, Mn and Si, get the results with the 60-degree mechanism only.

with h5py.File("../Fe-X-60-only/Si_data.h5","r") as fl:
    diff_60_aa_Si = np.array(fl['diff_aa'])
    diff_60_ab_Si = np.array(fl['diff_ab'])
    diff_60_bb_Si = np.array(fl['diff_bb'])
    diff_60_bb_Si_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("../Fe-X-60-only/Cu_data.h5","r") as fl:
    diff_60_aa_Cu = np.array(fl['diff_aa'])
    diff_60_ab_Cu = np.array(fl['diff_ab'])
    diff_60_bb_Cu = np.array(fl['diff_bb'])
    diff_60_bb_Cu_non_local = np.array(fl['diff_bb_nl'])

with h5py.File("../Fe-X-60-only/Mn_data.h5","r") as fl:
    diff_60_aa_Mn = np.array(fl['diff_aa'])
    diff_60_ab_Mn = np.array(fl['diff_ab'])
    diff_60_bb_Mn = np.array(fl['diff_bb'])
    diff_60_bb_Mn_non_local = np.array(fl['diff_bb_nl'])
        
with h5py.File("../Fe-X-60-only/Ni_data.h5","r") as fl:
    diff_60_aa_Ni = np.array(fl['diff_aa'])
    diff_60_ab_Ni = np.array(fl['diff_ab'])
    diff_60_bb_Ni = np.array(fl['diff_bb'])
    diff_60_bb_Ni_non_local = np.array(fl['diff_bb_nl'])
```

```python
# read in Messina et. al.'s results
pdc_dat= {}
Xlist = ["Cr", "Cu", "Ni", "Si","Mn"]
for X in Xlist:
    dat=[]
    with open("../SCMF_Messina_et_al/PDC_ratio_%s.dat"%(X),"r") as fl:
        for line in fl:
            arr = line.split()
            dat.append(float(arr[1]))
    pdc_dat[X] = np.array(dat)
```

```python
colorGF = "tab:blue"
colorMessina = "tab:orange"

colorInf = "tab:green"

infAll = "tab:green"
inf60 = infAll

lwMessina = 5
lwGF = 3

plt.figure(figsize=(18,15))

plt.subplot(2,3,1)
plt.semilogy(1000/Temp, pdc_dat["Cr"],ls='--', color=colorMessina, lw=lwMessina)
plt.semilogy(1000/Temp, diff_ab_Cr/diff_bb_Cr_non_local, ls="--", color=colorGF, lw=lwGF)
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L_{X-Fe}}{c_xL_{Fe-Fe}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.5,"X = Cr", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,3,2)
plt.semilogy(1000/Temp, pdc_dat["Mn"],ls='--', color=colorMessina, lw=lwMessina)
plt.semilogy(1000/Temp, diff_ab_Mn/diff_bb_Mn_non_local, color=colorGF, lw=lwGF)
plt.semilogy(1000/Temp, diff_60_ab_Mn/diff_60_bb_Mn_non_local, color=colorGF, lw=lwGF, ls="--")
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.text(0.8,0.5,"X = Mn", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,3,3)
plt.semilogy(1000/Temp, pdc_dat["Cu"],ls='--', color=colorMessina, lw=lwMessina)
plt.semilogy(1000/Temp, diff_ab_Cu/diff_bb_Cu_non_local, color=colorGF, lw=lwGF)
plt.semilogy(1000/Temp, diff_60_ab_Cu/diff_60_bb_Cu_non_local, color=colorGF, lw=lwGF, ls="--")
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.text(0.8,0.6,"X = Cu", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(2,3,4)
plt.semilogy(1000/Temp, pdc_dat["Ni"],ls='--', color=colorMessina, lw=lwMessina)
plt.semilogy(1000/Temp, diff_ab_Ni/diff_bb_Ni_non_local, color=colorGF, lw=lwGF)
plt.semilogy(1000/Temp, diff_60_ab_Ni/diff_60_bb_Ni_non_local, color=colorGF, lw=lwGF, ls="--")
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.ylabel(r'$\mathrm{\frac{L_{X-Fe}}{c_xL_{Fe-Fe}}}$', rotation=0, labelpad = 60, fontsize = 40)
plt.text(0.8,0.6,"X = Ni", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.subplot(2,3,5)
plt.semilogy(1000/Temp, pdc_dat["Si"],ls='--', label="Messina et.al", color=colorMessina, lw=lwMessina)
plt.semilogy(1000/Temp, diff_ab_Si/diff_bb_Si_non_local, label="GF method (All mechanisms)", lw=lwGF)
plt.semilogy(1000/Temp, diff_60_ab_Si/diff_60_bb_Si_non_local, label="GF Method ("+r'$60^{\circ}$RT'+" only)",
            color=colorGF, lw=lwGF, ls="--")
plt.xlabel(r'$\mathrm{\frac{1000}{T}(K^{-1})}$', fontsize = 25)
plt.text(0.8,0.6,"X = Si", horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes,
        fontsize = 20)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)

plt.tight_layout()
plt.savefig("PDC_FeX.pdf", bbox_inches="tight", format="pdf")
# plt.savefig("PDC_FeX.png", bbox_inches="tight", format="png")
# plt.savefig("PDC_FeX.eps", bbox_inches="tight", format="eps")
```
